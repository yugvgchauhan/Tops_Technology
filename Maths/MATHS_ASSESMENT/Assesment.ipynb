```python
#9
import matplotlib.pyplot as plt

# Data
x = [3, 4, 5, 6]
y = [1.5, 2, 2.5, 3]

# Plot
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot')
plt.show()
```

```python
#10
import numpy as np
import matplotlib.pyplot as plt

# Create array
x = np.arange(0, 6, 0.3)

# Plot
plt.plot(x, x, 'ro', label='x')
plt.plot(x, x**2, 'bs', label='x²')
plt.plot(x, x**3, 'g', label='x³')
plt.plot(x, x**4, ':', label='x⁴')

# Axis limits
plt.xlim(0, 6)
plt.ylim(0, 125)

# Labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Polynomial Plots')
plt.show()
```

```python
#11
import matplotlib.pyplot as plt

# Data
height = [179, 155, 191, 152, 188, 177]
names = ['QA', 'WB', 'EC', 'RD', 'TE', 'YF']

# Bar plot
plt.bar(names, height)
plt.xlabel('Names')
plt.ylabel('Height')
plt.title('Height Comparison')
plt.show()
```

```python
#12
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(100000)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(x, bins=10,edgecolor='black')
plt.title("Histogram with 10 bins")

plt.subplot(1, 3, 2)
plt.hist(x, bins=20,edgecolor='black')
plt.title("Histogram with 20 bins")

plt.subplot(1, 3, 3)
plt.hist(x, bins=50,edgecolor='black')
plt.title("Histogram with 50 bins")

plt.tight_layout()
plt.show()
```
